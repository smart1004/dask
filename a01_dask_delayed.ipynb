```python
#dask_delayed
from dask import delayed, compute
import dask
```

```python
@delayed
def square(num):
    print("square fn:", num)
    print()
    return num * num

@delayed
def sum_list(args):
    print("sum_list fn:", args)
    return sum(args)

items = [1, 2, 3]

computation_graph = sum_list([square(i) for i in items])

computation_graph.visualize()
```

```python
print("Result", computation_graph.compute())
```
